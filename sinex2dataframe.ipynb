```python
import pandas as pd
```

```python
with open("/home/ajay/work/sinex/ABPO_SlantTD_2020003_0000_66241.SNX",'r') as f:
    lines = f.readlines()
```

```python
# Loading data from sinex file and cleaning the data structure
```

```python
heading = lines[0]
column_names = heading[0:274]
row1 = heading[274:]
row1 = row1.replace(' ', ',')
row1_trimmed = row1[:len(row1)-1]
listed_row1 = row1_trimmed.split(',')
data = []
data.append(listed_row1)
for i in range (1,len(lines)):
    line = lines[i]
    line = line.replace(' ',',')
    trimmed_line = line[:len(line)-1]
    listed_line = trimmed_line.split(',')
    data.append(listed_line)
```

```python
# Creating well ordered indices and column headers for the data
```

```python
indices = []
for i in range (len(data)):
    indices.append(i)
```

```python
columns = []
columns.append(column_names[0:4])
columns.append(column_names[19:22])
columns.append(column_names[33:40])
columns.append(column_names[44:62])
columns.append(column_names[67:76])
columns.append(column_names[89:98])
columns.append(column_names[107:116])
columns.append(column_names[125:134])
columns.append(column_names[140:148])
columns.append(column_names[161:170])
columns.append(column_names[179:188])
columns.append(column_names[197:206])
columns.append(column_names[208:220])
columns.append(column_names[226:238])
columns.append(column_names[244:256])
columns.append(column_names[262:274])
```

```python
# Creating the dataframe
```

```python
df = pd.DataFrame(data,index = indices,columns = columns)
```
